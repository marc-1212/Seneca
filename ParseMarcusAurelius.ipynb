```python
import fitz
doc = fitz.open("Marco Aurelio-Meditaciones.pdf")
Final_Text = ""

pages = doc
pages = pages[44:209]
for page in pages:
    dictPage = page.get_text("dict")
    isFirsBlock = True
    for block in dictPage['blocks']:
      if isFirsBlock:
        isFirsBlock = False
        continue
      if "lines" in block:
        for line in block['lines']:
          if line['spans'][0]['font'] == "TimesNewRomanPSMT" and line['spans'][0]['size'] == 9.452380180358887:
            print(line)
            Final_Text += line['spans'][0]['text']
```

```python
doc[44].get_text("dict")
```

```python
Final_Text
```

```python
import re
pattern = r"\d+\.\s.*?(?=\d+\.\s)"

reflexions = re.findall(pattern, Final_Text, flags=re.DOTALL)
```

```python
clean_reflexions = []
for reflexion in reflexions:
  clean_reflexions.append(re.sub(r"^\d+\.\s", "", reflexion, flags=re.MULTILINE))
clean_reflexions
```
